```python
df_x = df.iloc[:,0:6]
df_y = df['Spectral Class']
```

```python
X = df_x
Y = df_y
dfX = pd.DataFrame(X, columns=df_x.columns)
print(dfX.head())
```

```python
len(Y.unique())
```

```python
knn = KNeighborsClassifier(n_neighbors=7,
weights='uniform',
algorithm='auto',
metric='euclidean')
 
X_train = X
y_train = Y
knn.fit(X_train, y_train)
```

```python
Data = [[6.2, 1.5, 4.2, 2.6]]
 
Y_pred = knn.predict(Data)
print("Hasil Prediksi : Jenis Bunga ", Y_pred)
```
